Q3. What is App routing in Flask? Why do we use app routes?


**App routing** in Flask is the process of mapping URLs to specific Python functions (view functions) using the `@app.route()` decorator. It allows users to access different parts of the web application by visiting defined URL paths. Routes are essential for organizing how URLs trigger specific actions, like displaying data or handling forms.

**Why use app routes?**
1. Define specific URLs for different functionalities.
2. Enable dynamic content by passing parameters in the URL.
3. Organize the application logically and maintainably.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

n Flask, the url_for() function is used for URL building. It generates the URL for a specific function in your Flask app, based on the function name and any arguments passed to it. This helps in avoiding hardcoding URLs in templates or view functions, making your application more flexible and maintainable.

```python
from flask import Flask, url_for, render_template_string

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the Home Page!'

@app.route('/about')
def about():
    return 'This is the About Page!'

@app.route('/contact')
def contact():
    return 'This is the Contact Page!'

@app.route('/links')
def links():
    home_url = url_for('home')
    about_url = url_for('about')
    contact_url = url_for('contact')

    return render_template_string('''
        <h1>Available Pages</h1>
        <ul>
            <li><a href="{{ home_url }}">Home</a></li>
            <li><a href="{{ about_url }}">About</a></li>
            <li><a href="{{ contact_url }}">Contact</a></li>
        </ul>
    ''', home_url=home_url, about_url=about_url, contact_url=contact_url)

if __name__ == '__main__':
    app.run(debug=True)
```
